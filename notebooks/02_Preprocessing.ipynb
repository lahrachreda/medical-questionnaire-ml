#### # Importation des bibliothèques nécessaires

```python
import pandas as pd
import sys
from pathlib import Path

```

#### # Chargement de Data

```python
df=pd.read_csv("../data/raw/Questionnaire.csv", sep=";",encoding="latin1")
```

#### # Configuration du projet

```python
# Définition de la racine du projet
PROJECT_ROOT = Path().resolve().parent

# Ajout de la racine au PYTHONPATH
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

print(f"Project Root : {PROJECT_ROOT}")
```

#### # Nettoyage des noms de colonnes

```python
from  ml.processing import clean_column_names
df = clean_column_names(df)
```

```python
from  ml.processing import clean_text_columns
df = clean_text_columns(df)
```
